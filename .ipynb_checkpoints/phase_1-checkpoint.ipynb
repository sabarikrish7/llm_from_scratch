```python
import os
import urllib.request

if not os.path.exists("the-verdict.txt"):
    url = "https://raw.githubusercontent.com/rasbt/LLMs-from-scratch/refs/heads/main/ch02/01_main-chapter-code/the-verdict.txt"
    file_path = "the-verdict.txt"
    urllib.request.urlretrieve(url, file_path)
```

```python
with open("the-verdict.txt",'r',encoding="utf-8") as f:
    raw_text=f.read()
    
```

```python
len(raw_text)
```

```python
import re



# Split by punctuation or whitespace, but keep the punctuation as separate tokens
result = re.split(r'([,.:;?_"!()\'\]|\-|\s])', raw_text)

# Clean up: remove empty tokens and strip spaces
result = [item.strip() for item in result if item.strip()]

preprocessed = result
print(preprocessed)
```

```python
len(preprocessed)
```

```python
##converting tokens to token ids
all_words=sorted(set(preprocessed))
vocab_size=len(all_words)
```

```python
vocab={char:id for id,char in enumerate(all_words)}
```

```python
print(vocab)
```

```python
class SImpleTokenizer():
    def __init__(self,vocab):
         self.str_to_int=vocab
         self.int_to_str={i:s for s,i in vocab.items()}
     def encode(text):
         result = re.split(r'([,.:;?_"!()\'\]|\-|\s])', text)
         result=[item.strip() for item in result if item.strip()]
         ids=[self.str_to_int[s] for s in result]
         return ids
         
```
